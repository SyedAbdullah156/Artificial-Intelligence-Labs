# Student Performance Analyzer

```python
num_std = int(input("Enter no of students: "))
records = dict()

for std in range(num_std):
    name = input(f"Enter name of student {std+1}: ")
    marks = []
    for i in range(3):
        marks.append(int(input(f"Enter marks for subject {i+1}: ")))
    records[name] = marks

print("\n" + "*" * 5 + " STUDENT PERFORMANCE " + "*" * 5 + "\n")

total_class_marks = 0
top_score = -1
top_student = ""

for name, marks in records.items():
    total = sum(marks)
    avg = total / len(marks)
    
    total_class_marks += avg
    
    if avg > top_score:
        top_score = avg
        top_student = name

    grade = "A" if avg >= 85 else "B" if avg >= 70 else "C" if avg >= 50 else "F"

    print(f"Name: {name}")
    print(f"Marks: {marks}")
    print(f"Average: {avg:.2f}")
    print(f"Grade: {grade}\n")

print("*" * 5 + " CLASS RECORD " + "*" * 5)
class_avg = total_class_marks / num_std
print(f"Class Average: {class_avg:.2f}")
print(f"Top Performer: {top_student} ({top_score:.2f}%)")
```

# Comment-Aware Word Frequency Analyzer

```python
freq = dict()
while True:
    line = input("Enter line: ")
    if line.startswith("#"):
        continue
    if line.strip().lower() == 'done':
        break
    words = line.strip().lower().split()
    for word in words:
        freq[word] = freq.get(word, 0) + 1

for word, count in freq.items():
    print(f"{word}: {count}")

print(freq[input("Enter word to check its count: ")])
```

# Shopping Cart & Billing System

```python
items = {"apple": 100, "banana": 50, "milk": 120, "bread": 80}
total_bill = 0
while True:
    name = input("Enter item name: ")
    quantity = int(input("Enter item quantity: "))
    if name not in items.keys():
        continue
    total_bill += items[name] * quantity
    if input("Type Done to finish: ").lower() == "done":
        break

discount = total_bill * 0.1 if total_bill >= 1000 else (total_bill * 0.05 if total_bill >= 500 else 0)

print(f"Original Bill: {total_bill}")
print(f"Discount Applied: {discount}")
print(f"Final Payable Amount: {total_bill - discount}")
```

# Data Cleaning & Statistical Summary

```python
readings = [23, 45, -999, 30, -999, 50, 41]
cleaned_readings = [x for x in readings if x !=  -999]
count = len(readings) - len(cleaned_readings)
print("Cleaned List: ", cleaned_readings)
print("Minimum Value:", min(cleaned_readings))
print("Maximum Value:", max(cleaned_readings))
print("Average Value:", sum(cleaned_readings)/len(cleaned_readings))
print("Count of faulty Values:", count)
```

# User Login Validation System

```python
users = {
    "admin": "1234",
    "ali": "pass123",
    "ahmad": "abcd"
}

for _ in range(3):
    username = input("Enter username: ")
    password = input("Enter password: ")
    if not username in users.keys(): 
        print("Invalid username")
    elif users[username] != password:
        print("Wrong password")
    else:
        print("Login successful")
        break
else:
    print("Account locked")
```
