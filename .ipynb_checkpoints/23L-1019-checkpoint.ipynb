```python
# Sum two numbers using typecast
num_int = 123
num_str = "456"
print("Data type of num_int:",type(num_int))
print("Data type of num_str before Type Casting:",type(num_str))
num_str = int(num_str)

print("Data type of num_str after Type Casting:",type(num_str))
num_sum = num_int + num_str
print("Sum of num_int and num_str:",num_sum)
print("Data type of the sum:",type(num_sum))
```

```python
# When We See a New key in dictionary
counts = dict()
names = ['Ali', 'Ahmed', 'Ali', 'Zain', 'Ahmed']
for name in names :
    if name not in counts:
        counts[name] = 1
    else :
        counts[name] = counts[name] + 1
print(counts)
name = input("Enter Name: ")
if name in counts:
    x = counts[name]
else :
    x = 0
print(name, "exists", x, "time(s).")
```

```python
# The pattern of checking to see if a key is already in a dictionary
# and assigning a default value
# Similar to upper code
names = ['Ali', 'Ahmed', 'Ali', 'Zain', 'Ahmed']
counts = {}
for name in names :
    counts[name] = counts.get(name, 0) + 1

print(counts)
name = input("Enter Name: ")
x = counts.get(name, 0)
print(name, "exists", x, "time(s).")

# Counting words in input
counts = dict()
line = input('Enter a line of text:')
words = line.split() # split() function considers space as delimiter by default;

# to use something else as delimiter,
# you have to pass that as parameter like line.split(';')

print('Words:', words)
print('Counting...')
for word in words:
    counts[word] = counts.get(word,0) + 1
print('Counts', counts)
```

```python
# Retrieving lists of Keys and Values
dict1 = { 'Ali' : 1 , 'Ahmad' : 42, 'Zain': 100}
print(dict1)
print(list(dict1))
print(list(dict1.keys()))
print(list(dict1.values()))
print(list(dict1.items()))
```

```python
# Definite Loops and Dictionaries
counts = { 'Ali' : 1 , 'Ahmad' : 42, 'Zain': 100}
for key in counts:
    print(key, counts[key])
for key,value in counts.items() :
    print(key, value)
```
