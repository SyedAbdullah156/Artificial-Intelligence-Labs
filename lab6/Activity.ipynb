# Game Node

```python
class GameNode:
    def __init__(self, value=None, children=[]):
        self.value = value
        self.children = children
```

# Minmax Algorithm

```python
# Minimax Algorithm (Recap)
def minimax(node, is_max):
    if not node.children:
        return node.value
    
    if is_max:
        best = float('-inf')
        for child in node.children:
            value = minimax(child, False)
            best = max(best, value)
        return best
    else:
        best = float('inf')
        for child in node.children:
            value = minimax(child, True)
            best = min(best, value)
        return best

leaf1 = GameNode(3)
leaf2 = GameNode(5)
leaf3 = GameNode(2)
leaf4 = GameNode(9)

min1 = GameNode(children=[leaf1, leaf2])
min2 = GameNode(children=[leaf3, leaf4])
root = GameNode(children=[min1, min2])
print("Optimal Value:", minimax(root, True))
```

# Minimax with Alpha-Beta Pruning

```python
# Minimax with Alpha–Beta Pruning
def minimax_alpha_beta(node, depth, alpha, beta, is_max):
    if not node.children:
        return node.value
    
    if is_max:
        best = float('-inf')
        for child in node.children:
            value = minimax_alpha_beta(child, depth+1, alpha, beta, False)
            best = max(best, value)
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return best
    else:
        best = float('inf')
        for child in node.children:
            value = minimax_alpha_beta(child, depth+1, alpha, beta, True)
            best = min(best, value)
            beta = min(beta, best)
            if beta <= alpha:
                break
        return best

print("Optimal Value with Alpha-Beta:", minimax_alpha_beta(root, 0, float('-inf'), float('inf'), True))
```

# Hill Climbing Algorithm - Finding maximum value in a landscape

```python
def hill_climbing(problem, start):
    current = start
    while True:
        neighbors = problem[current]

        # stop if no neighbors exist
        if not neighbors:
            return current
        
        best_neighbor = max(neighbors)

        if best_neighbor <= current:
            return current
        
        current = best_neighbor

landscape = {
    2: [3, 4],
    3: [5, 6],
    4: [6],
    5: [7],
    6: [7],
    7: []
}
print("Local Maximum:", hill_climbing(landscape, 2))
```

### Task 1 Analysis:

2) Basic minimax evaluates every leaf because it does not prune any branches. So it evaluates all 8 leaf nodes.
3) During alpha-beta evaluation (left → right) all leaf nodes are evaluated except 9. Node 9 is pruned. So in total 7 lead nodes are evaluated.

## Task-3, 4: Hill Climbing for Maximum Value

```python
import random

# Task 3: State Landscape
landscape = {
    1: [2, 3], 
    2: [4, 5], 
    3: [5], 
    4: [7], 
    5: [6], 
    6: [8], 
    7: [], 
    8: []
}

print("Task 3: Single Hill Climbing")
final_state = hill_climbing(landscape, 1)
print(f"Final local maximum reached: {final_state}")

# Task 4: Random Restart Hill Climbing
print("\nTask 4: Random Restart Hill Climbing")
runs = 5
best_overall = float('-inf')
states_list = list(landscape.keys())

for i in range(runs):
    start_state = random.choice(states_list)

    local_max = hill_climbing(landscape, start_state)
    
    print(f"Run {i+1}: Started at {start_state}, Local Max obtained = {local_max}")
    if local_max > best_overall:
        best_overall = local_max

print(f"\nTotal Restarts: {runs}")
print(f"Best solution among all runs: {best_overall}")
```

### Task 3: Hill Climbing
1) Step 1: Current state: 1. Neighbors: [2, 3]. Best: 3. Move to 3.
2) Step 2: Current state: 3. Neighbors: [5]. Best: 5. Move to 5.
3) Step 3: Current state: 5. Neighbors: [6]. Best: 6. Move to 6.
4) Step 4: Current state: 6. Neighbors: [8]. Best: 8. Move to 8.
5) Step 5: Current state: 8. Neighbors: []. Stop.

The algorithm stops at state 8 because it has no further neighbors to evaluate.

### Task 4 Explanation:
Standard Hill Climbing can get stuck in a "Local Maximum" (a state better than neighbors but not the global best). By running the algorithm multiple times from random starting points, it increases the chances of finding the true global optimal state.

# Task-5. Hill Climbing for Function Optimization

```python
def f(x):
    return -(x - 3)**2 + 9

def hill_climb_function(start):
    current = start
    while True:
        neighbors = [current + 1, current - 1]
        neighbor_values = [f(n) for n in neighbors]

        best_value = max(neighbor_values)
        best_neighbor = neighbors[neighbor_values.index(best_value)]

        print(f"Current x={current}, f(x)={f(current)}, Neighbors: {list(zip(neighbors, neighbor_values))}, Best={best_neighbor}")

        if f(current) >= best_value:
            return current

        current = best_neighbor

# Run
print("Task 5: Hill Climbing for Function Optimization")
final_x = hill_climb_function(0)
print(f"Final maximum value obtained at x={final_x}, f(x)={f(final_x)}")
```
