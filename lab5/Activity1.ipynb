```python
# Simple Custom Exception with super()
class WithdrawalError(Exception):
    """Raise Error if the withdraw amount is greater than balance"""
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(f"Insufficient balance! Tried to withdraw {self.amount}, but only {self.balance} is available")
    
def withdraw(balance, amount):
    if amount > balance:
        raise WithdrawalError(balance, amount)
    return balance - amount

try:
    print("New Balance:", withdraw(100, 50))
    print("New Balance:", withdraw(100, 150))
except WithdrawalError as e:
    print("Error:", e)
```

```python
# Advanced OOP Concepts - Polymorphism
class Animal: # Parent Class
    def speak(self):
        return "Some generic animal sound"
        
class Dog (Animal): # Child Class
    def speak(self): # Overriding speak()
        return "Bark"

class Cat (Animal): # Child Class
    def speak(self): # Overriding speak()
        return "Meow"
        
# creating objects
dog = Dog()
cat = Cat()
print(dog.speak())
print(cat.speak())
```

```python
# Example
# Base Class
class Animal:
    def __init__(self, name):
        self.name = name
        print(f"Animal Constructor Called for {self.name}")
    def eat(self):
        print(f"{self.name} is eating.")
    def speak(self):
        print('Generic Sound')
        
# Derived Class from Animal
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name) # Call parent constructor
        self.breed = breed
        print(f"Dog Constructor Called for {self.name}")
    def speak(self):
        print('Dog is barking.')

# Derived Class from Animal
class Fish(Animal):
    def __init__(self, name, water_type):
        Animal.__init__(self, name)
        self.water_type = water_type
        print(f"Fish Constructor Called for {self.name}")
    def swim(self):
        print(f"{self.name} is swimming in {self.water_type} water.")
    def speak(self):
        print('Fish are generally silent')

# Multilevel Inheritance (Dolphin -> Fish -> Animal)
class Dolphin(Fish):
    def __init__(self, name, water_type, intelligence_level):
        super().__init__(name, water_type)
        self.intelligence_level = intelligence_level
        print(f"Dolphin Constructor Called for {self.name}")
    def jump(self):
        print(f"{self.name} is jumping out of the water!")
    def speak(self):
        print('Dolphins are whistling')

# Creating Objects
print("----- Animal Object -----")
animal1 = Animal("Generic Animal")
print("\n----- Dog Object -----")
dog1 = Dog("Buddy", "Golden Retriever")

print("\n----- Fish Object -----")
fish1 = Fish("Nemo", "Fresh")
print("\n----- Dolphin Object -----")
dolphin1 = Dolphin("Flipper", "Salt", "High")
# Calling functions for each object
print("\n----- Animal Object Function -----")
animal1.speak()
print("\n----- Dog Object Function -----")
dog1.speak()
print("\n----- Fish Object Functions -----")
fish1.speak()
print("\n----- Dolphin Object Functions -----")
dolphin1.speak()
```

```python
# public_var
# _protected_var - single underscore is just the convention,
# it does not stop being access outside the classes
# __private_var

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner # public variable
        self.__balance = balance # private variable
    
# Creating an object
account = BankAccount("Alice", 1000)
# Trying to access the private attribute
print(account.owner) # works fine
print(account.__balance) # AttributeError
```

```python
# Getters & Setters Functions
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner # public variable
        self.__balance = balance # private variable
        
    def get_balance(self): # Getter Method
        return self.__balance

    def set_balance(self, amount): # Setter Method
        if amount < 0:
            print("Balance cannot be negative!")
        else:
            self.__balance = amount
    
# Creating an object
account = BankAccount("Alice", 1000)
# Accessing balance using getter
print(account.get_balance())
# Updating balance using setter
account.set_balance(1500)
# Again accessing balance using getter
print(account.get_balance())
```

```python
# Advanced OOP Concepts - Abstraction
from abc import ABC, abstractmethod

# Abstract Class
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass # No implementation, subclasses must define this method
        
    @abstractmethod
    def stop(self):
        pass # No implementation, subclasses must define this method

# Concrete Class (inherits from abstract class)
class car(Vehicle):
    def start(self):
        print("Car is starting...")
    def stop(self):
        print("Car is stoping...")
# error
# my_vehical = Vehicle()

my_car = car() # concrete class can be instantiated while abstract class can't
my_car.start()
my_car.stop()
```

```python
# Example
from abc import ABC, abstractmethod

# Abstract Class
class BankAccount(ABC):
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
    
    @abstractmethod
    def deposit(self, amount):
        pass # No implementation, subclasses must define this method

    @abstractmethod
    def withdraw(self, amount):
        pass # No implementation, subclasses must define this method
        
# Concrete Class
class SavingAccount(BankAccount):
    def deposit(self, amount):
        if amount < 0:
            print("Cannot deposit negative amount.")
        else:
            self.balance += amount
            print(f"Deposited ${amount}. New Balance: ${self.balance}")
            
    def withdraw(self, amount):
        if amount < 0:
            print ("Cannot withdraw negative amount.")
        elif amount > self.balance:
            print("Insufficient Funds!")
        else:
            self.balance -= amount
            print(f"Withdrawn ${amount}. New Balance: ${self.balance}")
    
# Creating an object
account = SavingAccount('Ali', 1000)
account.deposit(500)
account.withdraw(200)
account.withdraw(2000)
```

```python
# Exercise: Implement a banking system with classes Bank, Customer and Account
class Account:
    def __init__(self, balance=0):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    def deposit(self, amt):
        self.__balance += amt
    def withdraw(self, amt):
        if self.__balance >= amt:
            self.__balance -= amt
            print(f'Remaining Balance: {self.__balance}')
        else:
            print('Insufficient funds')

class Customer:
    def __init__(self, id, name, add):
        self.__id = id
        self.name = name
        self.address = add
        self.account = Account()
    def get_name(self):
        return self.name
    def get_address(self):
        return self.address
    def set_address(self, add):
        self.address = add
        print(f'Address updated to {add}')
    # Method chaining that hides 'account' class helps in creating abstraction user doesn't know about 'account' class used in customer class
    def deposit(self, amt):
        self.account.deposit(amt)
    def withdraw(self, amt):
        self.account.withdraw(amt)
    def get_balance(self):
        return self.account.get_balance()

class Bank:
    def __init__(self, bn):
        self.bank_name = bn
        self.customers = {}
    def add_customer(self, id, n, add):
        self.customers[id] = Customer(id, n, add)
    def get_customer(self, id):
        return self.customers.get(id, 'None')

bank = Bank('ABC Bank')
bank.add_customer(1, 'Ben', 'XYZ Street')
current_customer = bank.get_customer(1)
current_customer.deposit(200)
print(current_customer.get_balance())
current_customer.deposit(100)
print(current_customer.get_balance())

current_customer.withdraw(200)
print(current_customer.get_balance())
current_customer.withdraw(150)
print(current_customer.get_balance())
```
