# Top-N Items Analyzer

```python
sales = {
    "apple": 120,
    "banana": 85,
    "milk": 200,
    "bread": 150,
    "eggs": 90 
}
sales_list = list(sales.items())
sales_list_sorted = sorted(sales.items(), key=lambda sale: sale[1], reverse=True)
print(sales_list_sorted)
print("\n".join(f"{item}: {quantity}" for item, quantity in sales_list_sorted[:int(input("Enter number: "))]))
```

# Unique Student Registration System (Sets)

```python
unq_names = set()
while True:
    name = input("Enter Name: ")
    if name.lower() == 'done':
        break
    unq_names.add(name.strip())
print(f"Total number of students {len(unq_names)}")
print("\n".join(name for name in unq_names))
rm_name = input("Enter name to remove: ")
try:
    unq_names.remove(rm_name)
    print("\n".join(name for name in unq_names))
except e:
    print("Student not found")
```

# Number Analysis Using Functions & List Comprehension

```python
def compute_counts(numbers):
    return (len([number for number in numbers if number > 0]), len([number for number in numbers if number < 0]), len([number for number in numbers if number == 0]))

numbers = [int(number.strip()) for number in input("Enter numbers separated by spaces: ").split()]
(pos_count, neg_count, zer_count) = compute_counts(numbers)
print(f"Positive Count: {pos_count}")
print(f"Negative Count: {neg_count}")
print(f"Zero Count: {zer_count}")

sq_numbers = [number ** 2 for number in numbers if number > 0]
print(sq_numbers)
```
