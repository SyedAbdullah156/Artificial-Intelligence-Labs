```python
# List Modification Examples - List are Mutable
t = [9, 41, 12, 3, 74, 15]
t[3] = 10
print(t) # [9, 41, 12, 10, 74, 15]
t[:3] = [1, 2, 3]
print(t) # [1, 2, 3, 10, 74, 15]
```

```python
# List Comprehensions
# List Comprehension Example 1
squares1 = []
for n in range(10) :
    squares1.append(n**2)
squares2 = [n**2 for n in range(10)]
print("Square 1 List:", squares1)
print("Square 2 List:", squares2)
```

```python
# List Comprehension Example 2
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn']
short_planets1 = []
for planet in planets :
    if len(planet) < 6:
        short_planets1.append(planet.upper() + '!')
short_planets2 = [planet.upper() + '!' for planet in planets if len(planet) < 6]
print("Short Planet 1 List:", short_planets1)
print("Short Planet 2 List:", short_planets2)
```

```python
def count_negatives1(nums):
    negative_nums = []
    for num in nums :
        if num < 0 :
            negative_nums.append(num)
    return len(negative_nums)
def count_negatives2(nums):
    return len([num for num in nums if num < 0])
print(count_negatives1([-5, -1, -2, 0, 3]))
print(count_negatives2([-5, -1, -2, 0, 3]))
# It can also be written as return sum([num < 0 for num in nums])
# as False + True + True + False + False equals to 2.
```

```python
# Tuples
# often used for functions that have multiple return values.
t = (12, 5, 75) # it can also be written as: t = 12, 5, 75
print(t)
print(type(t))
```

```python
# Tuples Are Like Lists
x = ('Ali', 'Ahmad', 'Mustafa')
print(x[2])
print(len(x))
y = (1, 9, 2)
print(y)
print(max(y))
for item in y:
    print(item)
```

```python
# Tuple Functions
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
print(thistuple.count(5)) # returns number of times an element exists
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
print(thistuple.index(8))# returns the index of first occurrence of an element
```

```python
# Tuples are Comparable
print((0, 1, 2) < (5, 1, 2))
print((0, 1, 2000000) < (0, 3, 4))
print(( 'Jones', 'Sally' ) < ('Jones', 'Sam'))
print(( 'Jones', 'Sally') > ('Adams', 'Sam'))
```

```python
# Tuples and Dictionaries - The items() method in dictionaries returns a list of (key,value) tuples
d = dict()
d['a'] = 2
d['b'] = 4
print(d.items())
for (k,v) in d.items():
    print(k, v)
```

```python
# Sorting Lists of Tuples
d = {'a':10, 'c':22, 'b':1 }
print(d.items())
print(sorted(d.items()))
for k, v in sorted(d.items()):
    print(k, v)
```

```python
# Sort by Values Instead of Key
c = {'a':10, 'b':1, 'c':22, 'd':37}
temp = dict()
for k,v in c.items():
    temp[v] = k
print(temp)
print(type(temp))
temp = sorted(temp.items())
print(temp)
print(type(temp))
```

```python
# The top 2 most common words in input
counts = dict()
line = input("Enter Text: ")
line = line.strip()
words = line.split()
for word in words :
    counts[word] = counts.get(word, 0) + 1
temp1 = dict()
for k, v in counts.items():
    temp1[v] = k
temp1 = sorted(temp1.items(), reverse=True)
for k,v in temp1[:2]:
    print(v, k)
```

```python
# Python Sets
# Set items can appear in a different order every time you use them,
# Items cannot be referred to by index or key.
# Once a set is created, you cannot change its items
# but you can remove items and add new items.
myset = {"apple", "banana", "cherry"}
print(myset)
print(len(myset))
print(type(myset))
```

```python
# Python Sets - Duplicate values will be ignored:
thisset = {"apple", "banana", "cherry", "apple"}
print(thisset)
thisset2 = {"apple", "banana", "cherry", True, 1, 2}
print(thisset2) # True and 1 is considered the same value
```

```python
# Set items can be of any data type:
set1 = {"apple", "banana", "cherry"}
set2 = {1, 5, 7, 9, 3}
set3 = {True, False, False}
set4 = {"abc", 34, True, 40, "male"}
print(set1)
print(set2)
print(set3)
print(set4)
```

```python
# Add an element to a set, using the add() method:
thisset = {"apple", "banana", "cherry"}
thisset.add("orange")
print(thisset)
```

```python
# Add elements of a set and list to set:
thisset = {"apple", "banana", "cherry"}
tropical = {"pineapple", "mango", "papaya"}
mylist = ["kiwi", "orange"]
thisset.update(tropical)
print(thisset)
thisset.update(mylist)
print(thisset)
```

```python
# Remove element by using the remove() method:
thisset = {"apple", "banana", "cherry"}
thisset.remove("banana")
# If the item to remove does not exist, remove() will raise an error.
print(thisset)
```

```python
# Remove element by using the discard() method:
thisset = {"apple", "banana", "cherry"}
thisset.discard("banana")
# If the item to remove does not exist, discard() will NOT raise an error.
print(thisset)
```

```python
# Remove a random item by using the pop() method:
thisset = {"apple", "banana", "cherry"}
print(thisset.pop())
# when using the pop() method, you do not know which item that gets removed.
print(thisset)
```

```python
# The clear() method empties the set:
thisset = {"apple", "banana", "cherry"}
thisset.clear()
print(thisset)
```
