# Node Representation
Search algorithms store nodes, not just states.
A Node contains:
1) state
2) parent
3) depth
4) path cost

```python
class Node:
    def __init__(self, state, parent=None, cost=0, depth=0):
        self.state = state
        self.parent = parent
        self.cost = cost
        self.depth = depth
    
def get_path(node):
    path = []
    while node:
        path.append(node.state)
        node = node.parent
    return path[::-1]
```

# Breadth First Search (BFS)
1) Explores level by level
2) Uses a Queue (FIFO)
3) Finds shortest path in unweighted graph
4) Complete
5) Optimal (for unweighted graphs)
6) High memory usage

```python
from collections import deque
def bfs(graph, start, goal):
    queue = deque([Node(start)])
    visited = set()

    while queue:
        current_node = queue.popleft()
        current_state = current_node.state
        
        if current_state == goal:
            return get_path(current_node)

        visited.add(current_state)

        for neighbor in graph[current_state]:
            if neighbor not in visited:
                child_node = Node(neighbor, current_node)
                queue.append(child_node)

    return None
```

# Depth First Search (DFS)
1) Explores as deep as possible
2) Uses Stack or Recursion
3) Memory efficient
4) Not optimal
5) May get stuck in infinite depth

```python
def dfs(graph, start, goal):
    stack = [Node(start)]
    visited = set()

    while stack:
        current_node = stack.pop()
        current_state = current_node.state
        
        if current_state == goal:
            return get_path(current_node)

        visited.add(current_state)
        for neighbor in graph[current_state]:
            if neighbor not in visited:
                child_node = Node(neighbor, current_node)
                stack.append(child_node)

    return None
```

# Uniform Cost Search (UCS)
1) Expands node with lowest path cost
2) Uses Priority Queue
3) Works for weighted graphs
4) Complete
5) Optimal
6) Expensive memory usage

```python
import heapq

def ucs(graph, start, goal):
    pq = []
    heapq.heappush(pq, (0, Node(start)))
    visited = dict()
    
    while pq:
        cost, current_node = heapq.heappop(pq)
        current_state = current_node.state

        if current_state == goal:
            return get_path(current_node), current_node.cost

        if current_state not in visited or cost < visited[current_state]:
            visited[current_state] = cost

            for neighbor, edge_cost in graph[current_state]:
                total_cost = cost + edge_cost
                child_node = Node(neighbor, current_node, total_cost)
                heapq.heappush(pq, (total_cost, child_node))
                
    return None
```

# Iterative Deepening Search (IDS)
1) IDS combines:
    1. Memory efficiency of DFS
    2. Optimality of BFS
2) It performs DFS with increasing depth limit:
    1. Depth = 0
    2. Depth = 1
    3. Depth = 2
    4. ... until goal found
3) Complete
4) Optimal (unweighted graph)
5) Medium memory

```python
def depth_limited_dfs(node, graph, goal, limit):
    if node.state == goal:
        return node

    if node.depth >= limit:
        return None
        
    for neighbor in graph[node.state]:
        child = Node(neighbor, node, depth=node.depth + 1)
        result = depth_limited_dfs(child, graph, goal, limit)
        if result:
            return result

    return None
    
def ids(graph, start, goal, max_depth=10):
    for depth in range(max_depth):
        root = Node(start)
        result = depth_limited_dfs(root, graph, goal, depth)
        if result:
            return get_path(result)
    return None
```

# Student Record Management System

```python
class Student:
    school_name = "ABC School"
    
    def __init__(self, name, age, grade):
        if not Student.validate_age(age):
            raise ValueError("Age must be greater than 0")
        self.name = name
        self.age = age
        self.grade = grade

    def introduce(self):
        return f"{self.name},{self.age},{self.grade}\n"

    @classmethod
    def set_school_name(cls, name):
        cls.school_name = name

    @staticmethod
    def validate_age(age):
        return age > 0

# Step 1: Create file using 'x'
try:
    with open("students.txt", "x") as file:
        print("File created successfully.")
except FileExistsError:
    print("File already exists.")
    
# Step 2: Create Student objects
s1 = Student("Umer", 20, "A")
s2 = Student("Ali", 19, "A+")
s3 = Student("Amir", 17, "A-")
students = [s1, s2, s3]

# Step 3: Append records using 'a'
with open("students.txt", 'a') as file:
    for student in students:
        file.write(student.introduce())

# Step 4: Read file line by line (no extra spaces)
count = 0
with open("students.txt", "r") as file:
    for line in file:
        print(line.strip())
        count += 1
print("Total students:", count)

# Step 5: Search for a name using find()
search_name = input("Enter Name to search: ")
with open("students.txt", "r") as file:
    content = file.read()
    if content.find(search_name) != -1:
        print(f"{search_name}, found in file")
    else:
        print(f"{search_name}, not found in file")
        
```

# Vehicle Management

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def move(self):
        print("Vehical is moving") 
        
class Car(Vehicle):
    def __init__(self, brand, speed, num_doors):
        super().__init__(brand, speed)
        self.num_doors = num_doors

    def honk(self):
        print("Car is honking")
        
class Bike(Vehicle):
    def __init__(self, brand, speed, engine_capacity):
        super().__init__(brand, speed)
        self.engine_capacity = engine_capacity

    def rev(self):
        print("Bike is reving")
        
class ElectricCar(Car):
    def __init__(self, brand, speed, num_doors, battery_capacity):
        super().__init__(brand, speed, num_doors)
        self.battery_capacity = battery_capacity

    def charge(self):
        print("ElectricCar is charging")

v1 = Vehicle("Honda", 250)
v1.move()

c1 = Car("Toyota", 220, 4)
c1.move()
c1.honk()

b1 = Bike("Benelli", 300, 800)
b1.move()
b1.rev()

ec1 = ElectricCar("MG Phev", 220, 4, 500)
ec1.move()
ec1.honk()
ec1.charge()
```

# Graph Search Simulator (BFS + DFS + IDS)
    
1) BFS finds shortest path first.
2) DFS might find a deeper path first (not guaranteed shortest).
3) IDS behaves like BFS in terms of shortest path guarantee.

In this graph, shortest path depth = 2 → all return same.


Memory usage (worst case):
1) BFS → O(b^d) biggest memory
2) DFS → O(bd) smallest memory
3) IDS → O(bd) similar to DFS

So we got;
1) Most memory: BFS
3) Least memory: DFS (and IDS close)

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': [] 
}

print("Using BFS: ", bfs(graph, 'A', 'F'))
print("Using DFS: ", dfs(graph, 'A', 'F'))
print("Using IDS: ", ids(graph, 'A', 'F'))
```

# Weighted Graph Search Using UCS (Priority Queue)
1) BFS minimizes number of edges (depth).
2) UCS minimizes total path cost.
3) It uses a priority queue, not a simple queue.
4) It guarantees the minimum-cost path, even if:
    1. One path has fewer edges but high cost
    2. Another path has more edges but low cost

```python
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2)],
    'C': [('D', 1)],
    'D': []
}
path, cost = ucs(graph, 'A', 'D')
print(f"Original Graph -> Path: {path}, Cost: {cost}")

graph = {
    'A': [('B', 3), ('C', 1)],
    'B': [('D', 2)],
    'C': [('D', 1)],
    'D': []
}
path, cost = ucs(graph, 'A', 'D')
print(f"Modified Graph -> Path: {path}, Cost: {cost}")
```
