# Task 1, 2: Genetic Algorithm for Expression Optimization

```python
import random

def evaluateExpression(x, y, z):
    return 6 * x**3 + 9 * y**2 + 90 * z - 25

def fitness(x, y, z):
    ans = evaluateExpression(x, y, z)
    if ans == 0:
        return 99999
    else:
        return abs(1 / ans)

def run(start=0.99, end=1.01):
    # Initial population of 1000 individuals
    solutions = []
    for _ in range(1000):
        solutions.append((
                        random.uniform(0, 10), 
                        random.uniform(0, 10), 
                        random.uniform(0, 10)
        ))

    for generation_count in range(10000):
        rankedSolutions = []
        for solution in solutions:
            rankedSolutions.append((fitness(solution[0], solution[1], solution[2]), solution))
            
        rankedSolutions.sort(reverse=True)
        best_fit = rankedSolutions[0][0]
        worst_fit = rankedSolutions[-1][0]

        # Calculate average fitness
        total_fit = sum(fit for fit, sol in rankedSolutions)
        avg_fit = total_fit / len(rankedSolutions)

        print(f"Generation {generation_count} -> Best: {best_fit:.5f}, Worst: {worst_fit:.5f}, Avg: {avg_fit:.5f}")

        print(f"=== Generation {generation_count} best solution ===")
        print(f"Values (x,y,z): {rankedSolutions[0][1]} | Fitness: {rankedSolutions[0][0]}")
        
        # Stop if we are close enough to the exact answer
        if rankedSolutions[0][0] > 999:
            break
            
        # Selection
        bestSolution = rankedSolutions[:100]
        variables = []
        for solution in bestSolution:
            variables.append(solution[1][0])
            variables.append(solution[1][1])
            variables.append(solution[1][2])
            
        # Crossover and Mutation
        newGeneration = []
        for _ in range(1000):
            x = random.choice(variables) * random.uniform(start, end)
            y = random.choice(variables) * random.uniform(start, end)
            z = random.choice(variables) * random.uniform(start, end)
            newGeneration.append((x, y, z))
            
        solutions = newGeneration
```

```python
run()
```

# Task 3: Mutation Impact Experiment

```python
run(0.9, 1.1)
```

Explanation of Impact:
1) 0.99 to 1.01 (1% Mutation): This is highly fine-tuned (exploitation). It requires more generations to reach the answer, but it converges very smoothly and accurately without jumping over the perfect target.
2) 0.9 to 1.1 (10% Mutation): This causes massive leaps in the numbers (exploration). While it might get closer to the general target area faster in early generations, it struggles to converge on the exact decimal points because the mutations are too violent, causing the best fitness to fluctuate unpredictably at the end.

# Task 4: GA for Maximizing a Function

```python
import random

def f(x):
    return -(x**2) + 10*x + 5

# Generate random population of x values
solutions = [random.uniform(-50, 50) for _ in range(100)]

for generation_count in range(1000):
    # Fitness is directly the output of f(x) because we want to maximize it
    rankedSolutions = [(f(x), x) for x in solutions]
    rankedSolutions.sort(reverse=True)
    
    best_fitness = rankedSolutions[0][0]
    best_x = rankedSolutions[0][1]
    
    print(f"Gen {generation_count} | Best x: {best_x:.4f} | Max f(x): {best_fitness:.4f}")
    
    # Mathematically the max is 30 at x=5. Stop when we get extremely close.
    if round(best_x, 3) == 5.000:
        print(f"Converged after {generation_count} generations!")
        break
        
    bestSolution = rankedSolutions[:20]
    best_parents = [x for fit, x in bestSolution]
    
    newGeneration = []
    for _ in range(100):
        # Select parent and mutate slightly
        parent = random.choice(best_parents)
        child = parent + random.uniform(-0.5, 0.5)
        newGeneration.append(child)
        
    solutions = newGeneration
```

# Task 5: Genetic Algorithm for the Knapsack Problem

1) Representation of chromosomes: We use a binary string array. Since there are 3 items (A, B, C), a chromosome looks like [1, 0, 1]. A 1 means the item is put in the bag, and a 0 means it is left out.
2) Fitness function: The fitness is the total value of the items in the bag. However, if the total weight of the items exceeds the 50-unit capacity, the fitness is immediately penalized and drops to 0 (because the bag breaks).
3) Selection: We rank the binary strings by their total value. Valid combinations with the highest value (like keeping items B and C for 220 value / 50 weight) are selected as parents.
4) Mutation: A mutation simply flips a random bit in the array. For example, changing [1, 0, 1] to [1, 1, 1], meaning we randomly decide to try adding or removing an item to see if it improves the bag.

# Task 6: 8-Queens Problem using Genetic Algorithm

```python
import random

def fitness(chromosome):
    attacking_pairs = 0
    # Check every pair of queens to see if they attack each other
    for i in range(len(chromosome)):
        for j in range(i + 1, len(chromosome)):
            # Same row or same diagonal
            if chromosome[i] == chromosome[j] or abs(chromosome[i] - chromosome[j]) == abs(i - j):
                attacking_pairs += 1
                
    # Max non-attacking pairs is 28
    return 28 - attacking_pairs

# Initial population
solutions = []
for _ in range(1000):
    solutions.append([random.randint(1, 8) for _ in range(8)])

for generation_count in range(10000):
    rankedSolutions = [(fitness(sol), sol) for sol in solutions]
    rankedSolutions.sort(reverse=True)
    
    best_fitness = rankedSolutions[0][0]
    best_chromosome = rankedSolutions[0][1]
    
    print(f"=== Generation {generation_count} ===")
    print(f"Best Chromosome: {best_chromosome} Fitness = {best_fitness}")
    
    if best_fitness == 28:
        print("\nSolution Found!")
        print(f"Final Board Array: {best_chromosome}")
        print(f"Generations required: {generation_count}")
        break
        
    bestSolution = rankedSolutions[:100]
    newGeneration = []
    
    for _ in range(1000):
        # Crossover
        parent1 = random.choice(bestSolution)[1]
        parent2 = random.choice(bestSolution)[1]
        split = random.randint(1, 6)
        child = parent1[:split] + parent2[split:]
        
        # Mutation (5% chance to move a random queen to a random row)
        if random.random() < 0.05: 
            child[random.randint(0, 7)] = random.randint(1, 8)
            
        newGeneration.append(child)
        
    solutions = newGeneration

# Print final board
print("\nFinal board configuration:")
for row in range(1, 9):
    board_row = ""
    for col in range(8):
        if best_chromosome[col] == row:
            board_row += "Q "
        else:
            board_row += ". "
    print(board_row)
```
