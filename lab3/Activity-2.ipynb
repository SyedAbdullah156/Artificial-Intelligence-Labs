# Email Domain Frequency Analyzer 

```python
fname = input("Enter file name: ")
usernames = {}
domains = {}
try:
    fhand = open(fname, "r")

    for line in fhand:
        if not line.startswith("From "):
            continue
        words = line.split()
        email = words[1]
        username = email.split('@')[0]
        domain = email.split('@')[1]
        usernames[username] = usernames.get(username, 0) + 1
        domains[domain] = domains.get(domain, 0) + 1

    print(*[f"Username: {username}, Count: {count}\n" for username, count in usernames.items()])
    print(*[f"Domain: {domain}, Count: {count}\n" for domain, count in domains.items()])
    freq_username = sorted(usernames.items(), key=lambda item: item[1], reverse=True)[0]
    freq_domain = sorted(domains.items(), key=lambda item: item[1], reverse=True)[0]
    print(f"Most frequent Username: {freq_username[0]}, Count: {freq_username[1]}")
    print(f"Most frequent Domain: {freq_domain[0]}, Count: {freq_domain[1]}")
except e: 
    print("File cannot be opened")
```

# Sensor Data Analysis Using NumPy 

```python
import numpy as np

data = np.array([[10, 20, 30], 
                [40, 50, 60], 
                [70, 80, 90]]) 

print("Shape: ", data.shape)
print("Dimensions: ", data.ndim)
print("Data type: ", data.dtype)
print("Total Elements: ", data.shape[0] * data.shape[1])

b = data[:2, -2:]
print("Extracted Sub Array: \n", b)
b[0, 0] = b[0, 0] + 3
print("Extracted Sub Array After Change: \n", b)
print("Original Array: \n", data)

data = np.add(data, 10)
data = np.multiply(data, 2)
data = np.sqrt(data)
print("Final Result: \n", data)

```

#  Country Dataset Analysis Using Pandas 

```python
import pandas as pd
df = pd.DataFrame({'Country': ['Pakistan', 'India', 'China', 'USA'], 
            'Population': [241, 1420, 1440, 331], 
            'Area': [881913, 3287263, 9596961, 9833517]})
print("Dataframe: \n", df)
print("\nFirst two rows of df: \n", df[:2])
print("\nShape of df: ", df.shape)
print("\nColumn Names: ", ", ".join(list(df.columns)))
df.info()

print("\nSorting by Population: \n", df.sort_values(by='Population', ascending=False))
df['Rank'] = df['Area'].rank(ascending=False)
print("\nRank by Area: \n", df)
print("\nTotal Population: ", df['Population'].sum())
print("\nCountry with maximum population: ", df.loc[df['Population'] == df['Population'].max(), 'Country'].iloc[0])
print("\nCountry with minimum area: ", df.loc[df['Area'] == df['Area'].min(), 'Country'].iloc[0])

# Changing the index to country
df = df.set_index('Country')
print("\nAfter setting country as index of dataframe: \n", df)
print("\nData of Pakistan: \n", df.loc['Pakistan'])
print("\nData of China: \n", df.iloc[2])
print("\nOnly Population Column as Pandas Series: \n", df.loc[:, 'Population'], "\n")

# Saving DataFrame
df.to_csv('df.csv')

# Reading csv
df = pd.read_csv('df.csv')
print(df.shape)
print(df.columns)
print(df.loc[:2])
```

# Data Visualization Dashboard (Matplotlib)

```python
import matplotlib.pyplot as plt

scores = [45, 55, 60, 65, 70, 75, 80, 85, 90, 95]
subjects = ['Math', 'CS', 'AI', 'Stats']
marks = [85, 90, 78, 88]
```

```python
plt.title("Scores Data")
plt.xlabel("Score Values")
plt.ylabel("Indexes of Array")
plt.plot(scores)
plt.grid()
plt.show()
```

```python
plt.title("Bar Chart")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.bar(subjects, marks)
plt.show()
```

```python
plt.title("Histogram")
plt.xlabel("Scores")
plt.ylabel("Occurances")
plt.hist(scores)
plt.show()
```

```python
plt.title("Box Plot")
plt.xlabel("Mean")
plt.ylabel("Scores")
plt.boxplot(scores)
plt.show()
```
